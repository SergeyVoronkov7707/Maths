```python
R = 2
t=np.linspace(-5,5,41)
x,y=np.meshgrid(t,t)
F=x**2+y**2-R**2

plt.figure()
plt.contour(x, y, F, [0], linewidths=2,colors='b')
plt.gcf().gca().axis('image')
plt.gcf().gca().grid(True)
plt.show()
```

```python
a = 4
b = 2 
t=np.linspace(-6,6,50)
x,y=np.meshgrid(t,t)
F=(x/a)**2+(y/b)**2-1

plt.figure()
plt.contour(x, y, F, [0], linewidths=2,colors='g')
plt.gcf().gca().axis('image')
plt.gcf().gca().grid(True)
plt.show()
```

```python
a =3
b = 5
t=np.linspace(-6,6,50)
x,y=np.meshgrid(t,t)
F= x**2/a**2 - y**2/b**2 - 1

plt.figure()
plt.contour(x, y, F, [0], linewidths=2,colors='g')
plt.gcf().gca().axis('image')
plt.gcf().gca().grid(True)
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)

X, Y = np.meshgrid(X, Y)

Z = 2*X + 3*Y + 10
Z2 = 2*X + 3*Y -10
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z2, color = 'red')

show()
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)

X, Y = np.meshgrid(X, Y)

Z = 2*X**2 + 3*Y**2+ X + Y + 10
Z2 = 5*X**2 - 7*Y**2- X - Y -10
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2, color = 'red')

show()
```

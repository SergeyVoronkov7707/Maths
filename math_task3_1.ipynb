```python
#y = (1-np.exp(x)-x)/x
t = np.linspace(-5, 6, 201)
x,y=np.meshgrid(t,t)
F = np.exp(x) + x*(1 - y) - 1
F1 = x**2 - 1 - y


plt.figure()
plt.contour(x, y, F, [0], linewidths=2,colors='b')
plt.contour(x, y, F1, [0], linewidths=2,colors='r')
plt.gcf().gca().axis('image')
plt.gcf().gca().grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) +x*(1-y) - 1)

x1, y1 =  fsolve(equations, (-2, 3))
x2, y2 =  fsolve(equations, (-1, -2))
x3, y3 =  fsolve(equations, (2, 4))
              
print (x1, y1)
print (x2, y2)
print (x3, y3)
```
